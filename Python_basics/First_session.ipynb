```python
# Arithmatic operator

a=4
b=5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)  # floor division
print(a%b)  # modulus (remender)
print(a**b)  # a to the power b

```

```python
# Comparison operatort

print("a is equal to b ",a==b)
print("a is not equal to b ",a!=b)
print("a is  equal to b ",a>=b)
print("a is  equal to b ",a<=b)
```
