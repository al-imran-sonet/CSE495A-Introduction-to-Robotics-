Loop
1.  for loop
2.  while loop

```python
for i in range(10):   # by default it will start from 0
  print(i)

for i in range(2,10):  # It will make a boundary
  print(i)

for i in range(2, 100, 2):  # range(start, end, step count)
  print(i)
```

```python
i =0
while i<=10:
  print(i)
  i+=1
```

```python
for i in range(0,10):
  for j in range(0,10):
    print(i ,",",j ,"  ", end="")
  print("")
```
